```python
import numpy as np
import pandas as pd


def read_communities(file_path):
    communities = []

    with open(file_path, 'r') as file:
        for line in file:
            community = [int(num) for num in line.strip().split(',') if num]
            communities.append(community)

    return communities

def read_community_importances(file_path):
    community_importances = []

    with open(file_path, 'r') as file:
        for line in file:
          community_importances.append(float(line.strip()))

    return community_importances

def most_important_community(communities, importances):

    sorted_indices = np.argsort(importances)[::-1]

    sorted_communities = []

    for index in sorted_indices:
      sorted_communities.append(communities[index])

    gene_names = []

    # Input the gene names file here
    with open("", 'r') as file:
        for line in file:
          gene_names.append(line.strip())

    ans = []

    for elem in sorted_communities[0]:
        ans.append(gene_names[elem])

    return ans
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def count_subnetwork_frequencies(df):

    nameDict = {}
    count = 1
    for item in df['Final Subnetwork']:
        if item not in nameDict.values():
            nameDict[count] = item
            count+=1

    freq = {}
    for item in df['Final Subnetwork']:
        if item in freq:
            freq[item] += 1
        else:
            freq[item] = 1

    return nameDict, freq

def generate_plot_for_subnetworks(freq, nameDict):
    colors = plt.cm.get_cmap('tab10', len(nameDict))

    print(nameDict)

    plt.figure(figsize=(12, 8))

    plot_gnn_subnet = plt.bar(range(len(freq)), freq.values(), color=[colors(i) for i in range(len(freq))])

    labels = [f"{i}: {nameDict[i]}" for i in range(1, len(nameDict) + 1)]
    plt.legend(plot_gnn_subnet, labels, bbox_to_anchor=(1.05, 1))

    plt.show()

result_gnn_subnet = count_subnetwork_frequencies(pd.read_csv('subnetworks/subnetworks_gnn_subnet.csv'))
result_mean_aggr = count_subnetwork_frequencies(pd.read_csv('subnetworks/subnetworks_modified_alg_mean.csv'))
result_quantile_aggr = count_subnetwork_frequencies(pd.read_csv('subnetworks/subnetworks_modified_alg_q05.csv'))
```

```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def show_subnetwork_graph(subnetwork):
    path_to_PPI = "../TCGA/KIDNEY_RANDOM_PPI.txt"

    network = pd.read_csv(path_to_PPI, sep = ' ')

    protein_df = network[network['protein1'].apply(lambda x: x in subnetwork) & network['protein2'].apply(lambda x: x in subnetwork)]

    percentile_edge_score = np.percentile(network['combined_score'], q=95)

    # Create the graph
    G = nx.Graph()

    # Add edges with importance as weight
    for index, row in protein_df.iterrows():
        if row['combined_score'] > percentile_edge_score:
            # print(row['combined_score'])
            G.add_edge(row['protein1'], row['protein2'], weight=row['combined_score'], color='lavender')

    percentile_neighbour_count = np.percentile([len(list(nx.neighbors(G, node))) for node in G.nodes], q=90)
    node_colors = []

    edge_colors = {}
    for edge in G.edges:
        edge_colors[edge] = 'lavender'

    for node in G.nodes:
        neighbours = list(nx.neighbors(G, node))
        if len(neighbours) > percentile_neighbour_count:
            node_colors.append('limegreen')
            # print(G.edges(node))
            for edge in G.edges(node):
                if edge in edge_colors.keys():
                    edge_colors[edge] = 'lightgreen'
                else:
                    edge_colors[(edge[1], edge[0])] = 'lightgreen'
        else:
            node_colors.append('lightskyblue')

    print(G.edges)
    print(edge_colors)

    # Draw the graph
    pos = nx.spring_layout(G, k=1)  # positions for all nodes

    nx.draw(G, pos, with_labels=True, node_color=node_colors, edge_color=edge_colors.values())
    # nx.draw_networkx_edge_labels(G, pos)

    plt.show()
```
